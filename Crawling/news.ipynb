```python
from requests.compat import urlparse, urljoin
from requests.exceptions import HTTPError
import time
import sqlite3
from requests import Session
from bs4 import BeautifulSoup
import re
import datetime
```

```python
conn = sqlite3.connect('news.db')
cur = conn.cursor()
cur.executescript('''
    DROP TABLE IF EXISTS news;
    CREATE TABLE news(
        pk INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL,
        head TEXT NOT NULL,
        date TEXT NOT NULL,
        ref INTEGER NOT NULL
    );
''')
```

```python
url = 'https://news.naver.com/main/mainNews.nhn'
data = {
    'sid1':'',
    'firstLoad':'Y',
    'page':'',
}
session = Session()
# 100 : , 101 : , 102 : , 103 : /, 104 : , 105 : IT/
sid = [100, 101, 102, 103, 104, 105]
pagelist = [207, 356, 547, 70, 109, 70]




count = 0

while sid:
    sid1 = sid.pop(0)
    page = list(range(1,pagelist.pop(0)))
    data['sid1'] = sid1
    while page:
        try:
            count += 1
            data['page'] = page.pop(0)

            resp = session.request('POST',url, data=data)
            result = resp.json()['airsResult'].split('}')
            #time.sleep(0.02)


            for _ in result:
                try:
                    if re.search(r'"(?:title)":"([^s]+)',_).group(1):
                        cur.execute('''
                            INSERT INTO news(head, date, ref) VALUES(?,?,?)
                        ''', [re.search(r'"(?:title)":"([^s]+)',_).group(1), str(datetime.datetime.now()).split('.')[0], sid1])
                        conn.commit()
                except:
                    print('Error:{}'.format(count))

            if count%100 == 0:
                print(count)

        except:
            print('Error')
```
